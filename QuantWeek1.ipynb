```python
fdf["Daily Return"] = fdf.groupby(level=0)["Close"].pct_change()
```

```python
fdf["7-day  Moving Average"] = (
    fdf.groupby(level="Ticker")["Close"]
    .rolling(window=7)
    .mean()
    .reset_index(level=0, drop=True)
)
```

```python
fdf["30-day Moving Average"] = (
    fdf.groupby(level="Ticker")["Close"]
    .rolling(window=30)
    .mean()
    .reset_index(level=0, drop=True)
)
```

```python
fdf["Rolling Volatility(30d)"] = (
    fdf.groupby(level="Ticker")["Daily Return"]
    .rolling(window=30)
    .std()
    .reset_index(level=0, drop=True)
)
```
